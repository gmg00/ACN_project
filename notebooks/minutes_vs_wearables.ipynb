```python
%matplotlib qt
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '/Users/HP/Desktop/UNI/LM_1/ACN/ACN_project/data/address_to_wearables.json'
with open(file_path, "r") as json_file:
    address_to_wearables = json.load(json_file)
```

```python
address_to_num_wearables = {}
for address, wearables in address_to_wearables.items():
    address_to_num_wearables[address] = len(address_to_wearables[address])
```

```python
# Load dataset
df = pd.read_csv(f'/Users/HP/Desktop/UNI/LM_1/ACN/ACN_project/data/final_cleaned_df_2.csv')
```

```python
address_to_minutes = df['address'].value_counts().to_dict()
```

```python
file_path = '/Users/HP/Desktop/UNI/LM_1/ACN/ACN_project/data/layer3_edges.json'
with open(file_path, "r") as json_file:
    layer3_edges = json.load(json_file)
```

```python
unique_addresses = set()
for edge in layer3_edges:
    unique_addresses.add(edge[0])
    unique_addresses.add(edge[1])
```

```python
address_to_transactions = {}
for edge in layer3_edges:

    if edge[0] in address_to_transactions.keys():
        address_to_transactions[edge[0]] += 1
    else:
        address_to_transactions[edge[0]] = 1

    if edge[1] in address_to_transactions.keys():
        address_to_transactions[edge[1]] += 1
    else:
        address_to_transactions[edge[1]] = 1
```

```python
len(unique_addresses)
```

```python
print(f'len address_to_minutes: {len(address_to_minutes)}')
print(f'len address_to_wearables: {len(address_to_wearables)}')
```

```python
num_wearables = []
num_minutes = []

num_wearables_2 = []
num_minutes_2 = []
size = []
for address in address_to_num_wearables.keys():
    if address in list(unique_addresses):
        num_wearables_2.append(address_to_num_wearables[address])
        num_minutes_2.append(address_to_minutes[address])
        size.append(address_to_transactions[address])
    else:
        num_wearables.append(address_to_num_wearables[address])
        num_minutes.append(address_to_minutes[address])
        

num_wearables = np.array(num_wearables)
num_minutes = np.array(num_minutes)

num_wearables_2 = np.array(num_wearables_2)
num_minutes_2 = np.array(num_minutes_2)
```

```python
print(f'mean num wearables: {np.mean(num_wearables)}')
print(f'mean num minutes: {np.mean(num_minutes)}')
```

```python
plt.plot(num_minutes, num_wearables, 'k.', linestyle='', label=f'{len(num_minutes)}')
plt.scatter(num_minutes_2, num_wearables_2, s=size, alpha=0.7, edgecolors="black", color='red', zorder=1000)
plt.xlabel('Number of minutes', size=15)
plt.ylabel('Number of wearables', size=15)
plt.title('Number of wearables vs number of minutes', size=20)
plt.grid()
plt.legend(loc='best')
plt.show()
```

## Address to wearables

```python
wearables_occ = {}
collections_occ = {}

for address, wearables in address_to_wearables.items():
    for wearable in wearables:
        if wearable['nft_name'] in wearables_occ.keys():
            wearables_occ[wearable['nft_name']] += 1
        else:
            wearables_occ[wearable['nft_name']] = 1

        if wearable['nft_collection'] in collections_occ.keys():
            collections_occ[wearable['nft_collection']] += 1
        else:
            collections_occ[wearable['nft_collection']] = 1
```

```python
wearables_occ_sorted = dict(sorted(wearables_occ.items(), key=lambda item: item[1], reverse=True))
collections_occ_sorted = dict(sorted(collections_occ.items(), key=lambda item: item[1], reverse=True))
```
